```python
import os
from google import genai
from agents import Agent, Runner, function_tool, set_tracing_disabled
from agents.models.openai_chatcompletions import OpenAIChatCompletionsModel
from openai import AsyncOpenAI
import dotenv

dotenv.load_dotenv("/home/coder/eval-agents/.env")

set_tracing_disabled(True)  # no OpenAI key for tracing

genai_client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])

@function_tool
def google_search(query: str) -> str:
    """Search the web for current information."""
    response = genai_client.models.generate_content(
        model="gemini-2.5-flash",
        contents=query,
        config={"tools": [{"google_search": {}}]},
    )
    return response.text

# Gemini via OpenAI-compat endpoint
client = AsyncOpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key=os.environ["GOOGLE_API_KEY"]
)

agent = Agent(
    name="Assistant",
    instructions="Use google_search whenever you need current information.",
    tools=[google_search],
    model=OpenAIChatCompletionsModel(
        model="gemini-2.5-flash",
        openai_client=client
    )
)

# Runner.run_sync() cannot be used in Jupyter (event loop already running); use await instead
result = await Runner.run(agent, "What happened in AI this week?")
print(result.final_output)
```
